```python
import ast

def analyze_code(source):
    report = {
        "functions": [],
        "issues": []
    }

    try:
        tree = ast.parse(source)
    except SyntaxError as error:
        report["issues"].append(f"Syntax Error: {error}")
        return report

    for node in ast.walk(tree):
        if isinstance(node, ast.FunctionDef):
            report["functions"].append(node.name)
            if len(node.body) > 10:
                report["issues"].append(
                    f"Function '{node.name}' is too long"
                )

    if not report["functions"]:
        report["issues"].append("No functions found")

    return report


with open("sample_code.py") as file:
    code = file.read()

result = analyze_code(code)

print("Functions Found:", result["functions"])
print("Issues:")
for issue in result["issues"]:
    print("-", issue)
```
